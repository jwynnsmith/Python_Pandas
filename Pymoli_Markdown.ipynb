```python
# Create the bins in which Age Data will be held
bins = [0, 9, 14, 19, 24, 29, 34, 39, 44]

# Create the names for the eight bins
group_names = ["0-10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_bins = pd.cut(purchase_df["Age"], bins, labels=group_names)
age_bins
```

```python
purchase_df["Age Bins"] = pd.cut(
    purchase_df["Age"], bins, labels=group_names)
purchase_df.head()
```

```python
age_purchase_group_df = purchase_df.groupby(["Age Bins"])
age_purchase_group_df.head()
```

```python
age_ct_series = age_purchase_group_df["Price"].count()
age_ct_series
```

```python
age_avg_series = round(age_purchase_group_df["Price"].mean(), 2)
age_avg_series
```

```python
age_rev_series = round(age_purchase_group_df["Price"].sum(), 2)
age_rev_series
```

```python
# Normalized Total by Age
normalized_age_total = round(age_rev_series/age_ct_series, 2)
normalized_age_total
```

```python
age_df = pd.DataFrame({
    "Count": age_ct_series,
    "Normalized Totals": normalized_age_total,
    "Average Purchase Price": age_avg_series, 
    "Total Revenue":age_rev_series})
age_df
```

```python
# Do groupby descending values or

purchase_df["Spending Rank"] = purchase_df["Price"].sort_values(ascending=False)
purchase_df.head()
```

```python
purchase_rank_df_a = purchase_df.sort_values(by = "Spending Rank", ascending=False)
purchase_rank_df_b = purchase_rank_df_a[["SN", "Spending Rank"]]
purchase_rank_df_b.head(6)
```

```python
top_spend_ct = purchase_rank_df_b.head()["Spending Rank"].count()
top_spend_avg = round(purchase_rank_df_b.head()["Spending Rank"].mean(), 2)
top_spend_tot = purchase_rank_df_b.head()["Spending Rank"].sum()
```

```python
summary_top5_df = pd.DataFrame({
    "Number of Top Spenders":[top_spend_ct],
    "Average Purchase Price of Top 5 Spenders":[top_spend_avg],
    "Total Revenue of Top 5 Spenders":[top_spend_tot]
})
summary_top5_df
```

```python
item_purchase_group = purchase_df.loc[:,["Item Name", "Item ID", "Price", "Gender"]]
item_purchase_group.head()
```

```python
total_item_purchase = item_purchase_group.groupby(["Item ID", "Item Name", "Price"]).count()
total_item_purchase = total_item_purchase.rename(columns={"Gender":"Popularity Count"})
total_item_purchase = total_item_purchase.sort_values(by="Popularity Count", ascending=False)
total_item_purchase = total_item_purchase.reset_index(level=["Item ID", "Price"])
total_item_purchase["Total Purchase Value"] = total_item_purchase["Price"]*total_item_purchase["Popularity Count"]
total_item_purchase.head()
```

```python
item_profit_group = purchase_df.loc[:,["Item Name", "Item ID", "Price", "Gender"]]
item_profit_group.head()
```

```python
total_item_profit = item_profit_group.groupby(["Item ID", "Item Name", "Price"]).count()
total_item_profit = total_item_profit.rename(columns={"Gender":"Purchase Count"})
total_item_profit = total_item_profit.sort_values(by="Price", ascending=False)
total_item_profit = total_item_profit.reset_index(level=["Item ID", "Price"])
total_item_profit["Total Purchase Value"] = total_item_profit["Price"]*total_item_profit["Purchase Count"]
total_item_profit.head()
```
